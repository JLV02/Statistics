```python
# Imports and Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Histogram

```python
# Histogram plot
plt.hist(df['Current'], bins=5,
         alpha = 0.8,
         color = 'Orange',
         edgecolor = 'Black')

plt.title('Histogram of Current Values',
           font = 'Times New Roman',
           fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

# Normal Distribution

```python
# Normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

# x-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color = 'Green',
         linewidth=3)

plt.title('Histogram of Current Values',
           font = 'Times New Roman',
           fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
```

```python
# Histogram plot
plt.hist(df['Current'], bins=5,
         density=True,
         alpha = 0.8,
         color = 'Orange',
         edgecolor = 'Black')
         
plt.title('Normal Distribution of Current Values',
           font = 'Times New Roman',
           fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')

# Normal distribution curve
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color = 'Green',
         linewidth=3)
plt.show()
```

# Standard Normal Distribution N(0,1)

```python
# Standard Normal Distribution N(0,1)

# x-axis
x_min = df['Z-score'].min()
x_max = df['Z-score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof=1))

plt.plot(x, p,
         color = 'Blue',
         linewidth=3)

plt.title('Standard Normal Distribution of Current Values N(0,1)',
           font = 'Times New Roman',
           fontsize = 12)
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
```
